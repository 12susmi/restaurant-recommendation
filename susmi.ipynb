```python
from google.colab import files
uploaded = files.upload()
import pandas as pd
df = pd.read_csv("data_set.csv")
df.head()
```

```python
 df = df[[
    "Restaurant Name", "City", "Cuisines", "Price range",
    "Aggregate rating", "Rating text"
]]
df.fillna("", inplace=True)
df.head()
```

```python
df["combined_features"] = (
    df["City"] + " " +
    df["Cuisines"] + " " +
    df["Rating text"]
)
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words="english")
tfidf_matrix = tfidf.fit_transform(df["combined_features"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(tfidf_matrix)
```

```python
def recommend_restaurants(city, cuisine, price)
    user_pref = city + " " + cuisine
    user_vec = tfidf.transform([user_pref])
    scores = cosine_similarity(user_vec, tfidf_matrix).flatten()
    df["similarity_score"] = scores
    results = df[df["Price range"] == price]
    results = results.sort_values(by="similarity_score", ascending=False)
    return results[["Restaurant Name", "City", "Cuisines", "Price range", "Aggregate rating"]].head(10)
```

```python
recommend_restaurants("Makati City", "Japanese", 3)
```

```python
recommend_restaurants("Mandaluyong City", "Italian", 4)
```

```python

recommend_restaurants("", "Chinese", 2)
```
